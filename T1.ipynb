# Introduction

The device under investigation in this experiment is representative of any plant which uses thermodynamic principles to exchange heat and work continuously with the environment. All cyclic devices, from the steam cycle used in a power station, to the refrigeration cycle used in a domestic freezer must obey the same overall thermodynamic constraints. 

In this experiment, the cyclic heat and power plant is a commercially available heat pump. This heat pump uses a work input ($W$)  to pump heat ($Q_c$) from the cold source at temperature $T_c$ (in this case, air drawn over the evaporator with the aid of a fan), to a heat sink (in this case, hot water in a circulating system) at temperature $T_h$. As will be demonstrated, these air source heat pumps can reduce the amount of energy necessary for heating.

An ideal heat pump (Fig. 1) is equivalent to a Carnot cycle heat engine running in reverse. For the ideal reversible heat pump, a coefficient of performance (COP) can be defined as the ratio of what we want ($Q_h$) to what we put in ($W$). 
\begin{equation}
    \text{COP} = \frac{Q_h}{W} = \frac{Q_h}{Q_h - Q_c} 
\end{equation}
For an ideal Carnot cycle, for which $\frac{Q_h}{T_h} = \frac{Q_c}{T_c}$ this results in: 
\begin{equation}
    \text{COP} = \frac{T_h}{T_h - T_c}
\end{equation}
where the temperatures $T$ are absolute values in degrees kelvin. This represents the maximum heat $Q_h$ that can be provided per unit energy provided to the heat pump, and is a function of the source (cold) and sink (hot) temperatures.

## Energy balance on Condenser


Under adiabatic conditions in the water-condenser heat exchanger, the total energy transferred to the water is equal to: 
\begin{equation}
    Q_{h,c} = \dot{m}_w (h_{w,2}-h_{w,1}) = \dot{m}_w c_{p,w}(T_{2} - T_{1})
\end{equation}
where the mass flow rate $\dot{m}_w$ can be calculated using the equation.


\begin{equation}
    \dot{m}_w =  C_d \frac{\pi d_2^2}{4}\sqrt{\frac{2 \rho_w\Delta p}{1 - (d_2/d_1)^4}} = 9.578 \times 10^{-4} \sqrt{\Delta p (\text{bar}) \times 10^5}
\end{equation}

## Energy balance on evaporator

The total power input to the evaporator is equal to the energy supplied by the air: 
\begin{equation}
    Q_c = \dot{m}_a c_{p,a}(T_{4} - T_{3})
\end{equation}
and from energy conservation in the cycle, 
\begin{equation}
    Q_{h,a} = W + Q_c
\end{equation}

We now have two methods of calculating $Q_h$: from the difference in water temperatures across the condenser, $Q_{h,c}$, and from the temperature drop in the air flow  $Q_{h,a}$ . These may not agree, as the assumption of zero heat loss may not hold, and will lead to correspondingly different values of COP. 

The corresponding measured COP$_{h,c}$ and COP$_{h,a}$ can be compared with the ideal Carnot COP, by taking the hot temperature $T_h = T_2$ and $T_c = T_4$.
